## Decision Tree Classifier Implementation With Post Preprunning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
#df=sns.load_dataset('iris')
```

```python
#df.head()
```

```python
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

```python
df['target'] = dataset.target
```

```python
df.head()
```

```python
dataset.target
```

```python
#Independent and dependent features
x = df.iloc[:,:-1]
y = dataset.target
```

```python
x,y
```

```python
### train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
## Post Pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
x_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Post Pruning with max_depth =2
treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Prediction
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
report = classification_report(y_pred,y_test)
print(report)
```

```python
## Post Pruning with max_depth =2 with Entropy bydiffault it is ginni
treeclassifier = DecisionTreeClassifier(criterion='entropy',max_depth=2)
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
## Prediction
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
report = classification_report(y_pred,y_test)
print(report)
```
